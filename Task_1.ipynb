```python
import pandas as pd
import matplotlib.pyplot as plt

def show_historgram(file_path, file_name):
    data = pd.read_csv("{}/{}.txt".format(file_path,file_name),sep=';')
    emotions = data.iloc[:, 1]
    plt.figure(figsize=(10, 6))

    emotions.value_counts().plot(kind='bar', color='skyblue')
    emotion_counts = emotions.value_counts()
    plt.title('Emotion Categories Histogram For {} Dataset'.format(file_name))
    plt.xlabel('Emotion')
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    bars = plt.bar(emotion_counts.index, emotion_counts)
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, int(height), ha='center', va='bottom')

    plt.tight_layout()

    plt.show()

show_historgram("data","test")
show_historgram("data","train")
show_historgram("data","val")
show_historgram("data", "complete")
```

#### Comment:
The data is distributed super unbalanced. The training data should therefore be balanced before it is used to train the model.
